```python
import pandas as pd
```

```python
df = pd.read_csv('tiktok_google_play_reviews.csv')
```

```python
df = df.sample(5000)
```

```python
df['score']  = df['score'].astype(int)
```

```python
# INITIAL DATA  MANIPULATIONs
```

```python
# kucuk harfe donustur
df['content'] = df['content'].astype(str)
df["content"]=df["content"].str.lower()
# noktalama isaretlerini kaldirma
df["content"]=df["content"].str.replace("[^\w\s]","") 
# rakamlari kaldir
df["content"]=df["content"].str.replace("\d+","") 
# \newline ve enter\r kaldir
df["content"]=df["content"].str.replace("\n"," ").replace("\r","") 
# fazladan bosluklari azalt
df["content"]=df["content"].str.replace("  "," ")
```

```python
# strip out html tags
import html
def del_html(text):
   return html.unescape(text)

# Apply to multiple columns
df['content'] = df['content'].apply(del_html)
```

```python
df = df.dropna(subset=['content'])
```

```python
# LANGUAGE DETECTION
# There are comments in multiple languages in the comment section
# It is very difficult to simultaneously process different languages
# and apply sentiment analysis. Therefore, we filter out non-english
# comments in our analysis. 
```

```python
# Function to detect the language of a text using langdetect
import pandas as pd
from langdetect import detect

def detect_language(text):
    try:
        
        return detect(text)
    except:
        #there are emojis also, which may not be detected as language
        return 'unknown'

# Apply the language detection function to the 'text' column
df['language'] = df['content'].apply(detect_language)
df = df[df['language'] == 'en']
```

```python
# FIND ROOTS, REMOVE PREFIXES, CORRECT SPELLING MISTAKES
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from autocorrect import Speller
import re
from multiprocessing import Pool
import multiprocessing as mp

import concurrent.futures
from textblob import TextBlob

def find_word_roots(text):
    
    # Create a TextBlob object
    #blob = TextBlob(text)
    #text = blob.correct()
    spell = Speller(lang='en')
    # correct text
    text = spell(text)
    
    pr = PorterStemmer()
    lm = WordNetLemmatizer()
    
    # find roots
    words = text.split(" ");
    kokler = [pr.stem(word) for word in words]
    
    #remove prefixes 
    kokler_k = [];
    for kok in kokler :
        kok=spell(kok)
        #blob = TextBlob(kok)
        # Correct the text
        #kok = str(blob.correct())
        
        kokler_k.append( lm.lemmatize(kok,pos = 'v') )
        
    #remove     
    
    ret_str = "";
    for kok in kokler_k :
        ret_str = ret_str +" "+kok
        
    # remove https part
    ret_str =re.sub(r'https?\S+', '', ret_str)    
    print(ret_str)    
    return ret_str

#num_processes = mp.cpu_count()*1;
#num_processes = 1;
#with Pool(num_processes) as pool:
#    # Use the Pool's map function to apply your function to the column
#    results = pool.map(find_word_roots, df['text'])

# Add the results as a new column in your DataFrame
#df['textt'] = results


#with concurrent.futures.ProcessPoolExecutor() as executor:
#    results = list(executor.map(find_word_roots, df['text']))

df['content'] = df['content'].apply(find_word_roots)  
```

```python
# Remove STOP words
# Taken from Amazon 
import re
import nltk

from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize
from nltk.corpus import wordnet
from nltk.corpus import stopwords

#import nltk resources
resources = ["wordnet", "stopwords", "punkt", \
             "averaged_perceptron_tagger", "maxent_treebank_pos_tagger"]

for resource in resources:
    try:
        nltk.data.find("tokenizers/" + resource)
    except LookupError:
        nltk.download(resource)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import TreebankWordTokenizer

stop_words = stopwords.words("english")

stop_words = [word.replace("\'", "") for word in stop_words]
```

```python
# remove stopwords

import six
import time

i = 0;
nan_index = []
text_wo_sw = []
for text in df['content']:
    
    ret_str ="";
    if isinstance(text, six.string_types):
    
        splitted_text = text.split(" ");
        for word in splitted_text:
            #print(word)

            if len(word)>0 and isinstance(word, six.string_types):
                if word not in stop_words:
                    ret_str = ret_str +" "+word
                        
            #else :
                #print(word)
                #print('not string')
    else :
        #print(i) 
        #print(text)
        nan_index.append(i)
        #print('not a string unfortunately')
    
    text_wo_sw.append(ret_str)
    i=i+1;    
df['content'] =  text_wo_sw 
df.drop(df.index[nan_index], inplace=True)
```
